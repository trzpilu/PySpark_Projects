```
path_adult = "/databricks-datasets/adult/adult.data"

field_names = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "native-country",
    "label"
]

df_adult = spark.read.csv(path_adult)
df_adult = df_adult.toDF(*field_names)
df_adult.registerTempTable("adult")
```

```
%sql
SELECT DISTINCT(`education`) FROM adult
```

```
%sql
SELECT `label`, `education`, count(*) as count FROM adult
WHERE `education` = ' Doctorate' or `education` = ' Masters'
GROUP BY `label`, `education`
ORDER BY count DESC
```

```
%sql
SELECT DISTINCT(`race`) FROM adult
```

```
%sql
SELECT `label`, `race`, `sex`,count(*) as count FROM adult
GROUP BY `label`, `race`, `sex`
ORDER BY count DESC
```

```
%sql
SELECT `label`, `race`, `sex`,count(*) as count FROM adult
GROUP BY `label`, `race`, `sex`
ORDER BY count DESC
```

```
%sql
SELECT `label`, `race`, `marital-status`,count(*) as count FROM adult
GROUP BY `label`, `race`, `marital-status`
ORDER BY count DESC
```

```
%sql
SELECT `label`, `race`, `marital-status`,count(*) as count FROM adult
GROUP BY `label`, `race`, `marital-status`
ORDER BY count DESC
```
