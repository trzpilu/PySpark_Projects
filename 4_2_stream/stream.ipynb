```
entries_df = spark.readStream.format("rate")\
    .option("rowsPerSecond", 5)\
    .load()

entries_df.isStreaming
```

```

display(entries_df)
```

```
import pyspark.sql.functions as F

entries_df = entries_df.withColumn(
    "start",
    F.window(entries_df.timestamp, "10 seconds")["start"]
)

stream_sink_df = entries_df.groupBy(entries_df.start).count()
```

***Ważne do zapamiętania!***

Skoro ustawiliśmy: 

****entries_df = .option("rowsPerSecond", 5)\****, 

czyli tworzenie odczytu co 0,2 sekundy, ergo jeśli chcemy stworzyć limit elementów równy 50, to musimy ustawić timestamp na 10 sekund, skoro 5 elementów na sekundę * 10-sekundowy timestamp, daje maksymalnie 50 elementów. Stąd ustawiamy timestamp na 10 sekund: 

****F.window(entries_df.timestamp, "10 seconds")["start"]****.

Mniej otrzymamy przy zatrzymaniu lub starcie streamingu, gdyż odlicznie nieczęsto zaczyna się idealnie na początku 10 sekund timestampu.

```
query = stream_sink_df.writeStream.format("memory")\
    .queryName("test_groups")\
    .outputMode("complete")\
    .start()
```

```
%sql
select * from test_groups
```
