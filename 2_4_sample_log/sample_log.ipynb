```
rdd_real_logs = sc.textFile("/databricks-datasets/sample_logs/part-*")
rdd_real_logs.take(3)
```

```
import re

rdd_filtered = rdd_real_logs \
    .filter(lambda word: re.search("_3[0-9][0-9]", word)) 

rdd_filtered.take(10)
```

```
rdd_names_flat = rdd_filtered.flatMap(lambda text: text.split("/endpoint"))

rdd_names_flat.take(5)
```

```
rdd_names_flat_2 = rdd_names_flat.flatMap(lambda text: text.split(" HTTP"))
rdd_names_flat_2.take(5)
```

```
rdd_filtered_2 = rdd_names_flat_2 \
    .filter(lambda word: re.search("_3[0-9][0-9]", word)) 

rdd_filtered_2.take(10)
```

```
sorted(
    rdd_filtered_2.countByValue().items(),
    key=lambda el:el[1], reverse=True
)[:10]
```
