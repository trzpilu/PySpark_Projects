```
# Tworzymy obiekt RDD
```

```
rdd = sc.parallelize([
    "this is positive message",
    "this is negative comment",
    "very positive and happy"
])
```

```
# Piszemy funkcję manipulującą
```

```
def simple_sentiment(text):
    if "positive" in text:
        return "positive"
    else:
        return "negative"
```

```
# Używamy tej funkcji, wykonujemy redukcję i wyświetlamy wynik
```

```
def fun(a):
    return (simple_sentiment(a), 1)

rdd.map(fun).reduceByKey(lambda a,b: a+b).take(10)
```
