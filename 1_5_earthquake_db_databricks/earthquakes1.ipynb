```
# Import pandas for table
```

```
import pandas as pd
```

```
# Import data from earthquake_database.csv.
```

```
df = spark.read.csv("/FileStore/tables/earthquake_database.csv")
```

```
# Manipulate data to show that most powerful earthquakes originated in deepest regions of crust.
# To do so, we reorder the data by "Magnitude" (cell: _c8).
```

```
df_magnitude = df.orderBy('_c8', ascending=False)
```

```
# Present data for the top 10 greatest earthquakes.
```

```
# Make new list
```

```
top_10_earthquakes = df_magnitude.head(11)
print (top_10_earthquakes)
```

```
# Present top_10_earthquakes data in table format using pandas.
```

```
earthquake_table =  pd.DataFrame(top_10_earthquakes)
print(earthquake_table)
```

```
# Going back, we create a newdf by selecting columns _c0 (Date), _c5 (Depth) and _c8 (Magnitude), with df2 taking the top 10 entries.
```

```
newdf = df_magnitude[['_c0','_c5','_c8']]
df2 = newdf.head(11)
```

```
# Present df2 data in table format using pandas. One can clearly see that only one earthquake in the top 10 has a source below 5 km, while most originated around 30 km below the surface.
```

```
earthquake_table_df2 =  pd.DataFrame(df2)
print(earthquake_table_df2)
```
