```
rdd = spark.read.csv("/databricks-datasets/adult/adult.data").rdd
rdd.take(10)
```

```
rdd.getNumPartitions()
```

```
rdd_divorced = rdd \
    .filter(lambda row: "Divorced" in row._c5)
rdd_divorced.take(10)
```

```
rdd_sample_divorced = sc.parallelize(rdd_divorced.take(100))
rdd_sample_divorced.take(10)
```

```
rdd_sample_divorced_filtered = rdd_sample_divorced.sortBy(lambda row: row._c0, ascending=True)
rdd_sample_divorced_filtered.take(10)
```

```
rdd_divorced_managers = rdd_sample_divorced_filtered.map(lambda row: (row._c6, 1))
rdd_divorced_managers.take(10)
```

```
divorced_managers = rdd_divorced_managers.countByKey()[' Exec-managerial']
divorced = rdd_divorced_managers.count()
calculation = (divorced_managers/divorced)*100

text = "Of the {} divorcees sampled, {} held exec-managerial positions.\
\nTherefore {}% are of the divorcees held exec-managerial positions"\
.format(divorced, divorced_managers, calculation)

print(text)
```

```
rdd_filtered_divorced_managers = rdd_sample_divorced_filtered \
    .filter(lambda row: "Exec-managerial" in row._c6)
rdd_filtered_divorced_managers.getNumPartitions()
```

```
rdd_repartioned = rdd_filtered_divorced_managers.repartition(1)
rdd_repartioned.getNumPartitions()
```

```
rdd_repartioned.coalesce(1).saveAsTextFile("/FileStore/rdd/sample_managers_divorced_file")
```
